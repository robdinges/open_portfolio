```python
class Car:
    cars = []
    def __init__(self, make, year):
        self.make = make
        self.year = year
        Car.cars.append(self.make)

    def __repr__(self):
        return f'Mijn auto uit {self.year} is van het merk {self.make}.'
```

```python
car1 = Car('Renault',2013)
car2 = Car('Volkswagen', 1996)
```
