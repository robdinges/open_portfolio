```python
from OpenPortfolioLib import *
```

```python
from datetime import date
import logging

# Initialize logging
logging.basicConfig(level=logging.CRITICAL)

# Initialize the necessary objects
product_collection = ProductCollection()
transaction_manager = TransactionManager()

# Create a client
client = Client(client_id=1, name="John Doe")

# Add a portfolio to the client
portfolio = client.add_portfolio(portfolio_id=1)

# Add a cash account to the portfolio
portfolio.add_cash_account(account_id=123, start_balance=10000)

# Create products
product1 = Stock(product_id=1, description="ABC Corp", minimum_purchase_value=1, smallest_trading_unit=1, issue_currency='EUR')
product2 = Bond(instrument_id=2, description="XYZ Bond", minimum_purchase_value=1000, smallest_trading_unit=1, issue_currency='EUR', start_date=date(2023, 1, 1), maturity_date=date(2028, 1, 1), interest_rate=0.05, interest_payment_frequency=PaymentFrequency.YEAR)
product_collection.add_product(product1)
product_collection.add_product(product2)

product1.add_price(date(2023,1,1), 1.23)
product2.add_price(date(2023,1,1), 3.45)


# Function to print the balance of a given account
def print_account_balance(portfolio, account_id, currency='EUR', account_type=AccountType.CASH):
    account = portfolio.search_account_id(account_id, currency, account_type)
    if account:
        print(f"Balance for account {account_id}: {account.balance}")

# Create and execute a deposit transaction
deposit_transaction = Transaction(transaction_date=date(2024, 1, 5), portfolio_id=portfolio.portfolio_id, account_id=123)
cash_movement = CashMovement(
    transaction=deposit_transaction,
    amount_account_currency=5000,
    amount_original_currency=5000,
    movement_type=MovementType.DEPOSIT,
    transaction_number=deposit_transaction.transaction_number
)
deposit_transaction.add_cash_movement(cash_movement)
transaction_manager.execute_transaction(deposit_transaction, portfolio, product_collection)

# Create and execute a purchase transaction
purchase_transaction = transaction_manager.create_transaction(
    transaction_date=date(2024, 1, 10),
    portfolio_id=portfolio.portfolio_id,
    template=TransactionTemplate.BUY,
    product_id=product1.instrument_id,
    amount=10,
    price=100,
    account_id=123,
    cost=10
)
transaction_manager.execute_transaction(purchase_transaction, portfolio, product_collection)

# List the holdings in the portfolio after the transactions
portfolio.list_holdings()

# Create and execute a withdrawal transaction
withdrawal_transaction = Transaction(transaction_date=date(2024, 1, 20), portfolio_id=portfolio.portfolio_id, account_id=123)
cash_movement = CashMovement(
    transaction=withdrawal_transaction,
    amount_account_currency=-2000,
    amount_original_currency=-2000,
    movement_type=MovementType.WITHDRAWAL,
    transaction_number=withdrawal_transaction.transaction_number
)
withdrawal_transaction.add_cash_movement(cash_movement)
transaction_manager.execute_transaction(withdrawal_transaction, portfolio, product_collection)

# Create and execute a sale transaction
sale_transaction = transaction_manager.create_transaction(
    transaction_date=date(2024, 1, 25),
    portfolio_id=portfolio.portfolio_id,
    template=TransactionTemplate.SELL,
    product_id=product2.instrument_id,
    amount=5,
    price=110,
    account_id=123
)
transaction_manager.execute_transaction(sale_transaction, portfolio, product_collection)

# List the holdings in the portfolio after the transactions
portfolio.list_holdings()

# Display the transaction table
portfolio.transactions_table()

# Display the returns table
portfolio.returns_table(start_date=date(2024, 1, 1), end_date=date(2024, 12, 31))

product_collection.list_products()

```
