```python
from OpenPortfolioLib import *
```

```python
from datetime import date
import logging


time_travel = TimeTravel()

# Initialize logging
logging.basicConfig(level=logging.ERROR)

# Initialize the necessary objects
product_collection = ProductCollection()
transaction_manager = TransactionManager()

# Create a client
client = Client(client_id=1, name="John Doe")

# Add a portfolio to the client
portfolio = client.add_portfolio(portfolio_id=1, default_currency='EUR', cost_in_transaction_currency=True)
portfolio_analytics = PortfolioAnalytics(portfolio)

# Add a cash account to the portfolio
portfolio.add_cash_account(account_id=123, start_balance=10000)
portfolio.add_cash_account(account_id=123, account_type=AccountType.CASH, start_balance=5000, currency='USD')

currency_prices = CurrencyPrices()
currency_prices.add_price('USD', date(2022,1,1), 1.1)
currency_prices.add_price('EUR', date(2022,1,1), .9, 'USD')
```

```python
portfolio.list_accounts()
```

```python
# Product 1
product1 = Stock(product_id=1, description="ABC Corp", minimum_purchase_value=1, smallest_trading_unit=1, issue_currency='EUR')
product_collection.add_product(product1)

# Create and execute a buy transaction
transaction_manager.create_and_execute_transaction(
    transaction_date=date(2022, 1, 5),
    portfolio_id=1,
    template=TransactionTemplate.BUY,
    account_id=123,
    portfolio=portfolio,
    product_collection=product_collection,
    product_id=1,
    amount=2,
    price=3.5,
    #cost=10
)

# Create and execute a buy transaction
transaction_manager.create_and_execute_transaction(
    transaction_date=date(2022, 2, 20),
    portfolio_id=portfolio.portfolio_id,
    template=TransactionTemplate.SELL,
    account_id=123,
    portfolio=portfolio,
    product_collection=product_collection,
    product_id=1,
    amount=1,
    price=3.1,
    #cost=10
)

product_collection.list_products()
portfolio.list_holdings(time_travel, date(2022,12,31))

```

```python
# Product 2
product2 = Bond(instrument_id=2, description="XYZ Bond", minimum_purchase_value=1000, smallest_trading_unit=1000, issue_currency='EUR', start_date=date(2021, 1, 1), maturity_date=date(2028, 1, 1), interest_rate=0.1, interest_payment_frequency=PaymentFrequency.YEAR)
product_collection.add_product(product2)
product2.add_price(date(2022,1,1), 100.0)
product2.add_price(date(2022,2,1), 100.0)
product2.add_price(date(2022,3,1), 100.0)
product2.add_price(date(2022,4,1), 100.0)


transaction_manager.create_and_execute_transaction(
    transaction_date=date(2029, 5, 30),
    portfolio_id=portfolio.portfolio_id,
    template=TransactionTemplate.BUY,
    account_id=123,
    portfolio=portfolio,
    product_collection=product_collection,
    product_id=2,
    amount=1000,
    price=1.0,
    #cost=10
)

# Create and execute a buy transaction
transaction_manager.create_and_execute_transaction(
    transaction_date=date(2022, 6, 10),
    portfolio_id=portfolio.portfolio_id,
    template=TransactionTemplate.SELL,
    account_id=123,
    portfolio=portfolio,
    product_collection=product_collection,
    product_id=2,
    amount=1000,
    price=1.01,
    #cost=10
)

product_collection.list_products()
portfolio.list_holdings(time_travel, date(2022,12,31))

```
