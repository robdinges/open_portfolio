```python
import datetime
```

```python
class CashMutation:
    def __init__(self, date, amount):
        self.date = date
        self.amount = amount
```

```python
class InterestRateChange:
    def __init__(self, date, rate):
        self.date = date
        self.rate = rate
```

```python
class Portfolio:
    def __init__(self):
        self.bonds = []
        self.cash_mutations = []
        self.interest_rate_changes = []
        self.cash = 0
 
    def add_cash(self, date, amount):
        self.cash += amount
        self.cash_mutations.append(CashMutation(date, amount))
 
    def withdraw_cash(self, date, amount):
        self.cash -= amount
        self.cash_mutations.append(CashMutation(date, -amount))
 
    def change_interest_rate(self, date, rate):
        self.interest_rate_changes.append(InterestRateChange(date, rate))
 
    def calculate_cash_value(self, date):
        # Combine and sort all events by date
        events = []
        for mutation in self.cash_mutations:
            events.append((mutation.date, mutation.amount, None))
        for change in self.interest_rate_changes:
            events.append((change.date, None, change.rate))
        
        # Sort events based on the date
        events.sort()
        
        cash_value = 0
        current_rate = 0.0
        last_date = events[0][0] if events else date  # Start from the first event's date
        
        for event in events:
            event_date, amount, rate = event
            
            if event_date > date:
                break
            
            # Calculate interest up to the current event date
            days = (event_date - last_date).days
            if days > 0:
                cash_value += cash_value * current_rate * (days / 365)
            
            # Update cash value or rate based on the event type
            if amount is not None:
                cash_value += amount
            if rate is not None:
                current_rate = rate
            
            last_date = event_date
        
        # Apply interest from the last event date to the requested date
        days = (date - last_date).days
        if days > 0:
            cash_value += cash_value * current_rate * (days / 365)
        
        return cash_value
 
    def get_total_value(self, date):
        # Placeholder for bond value calculation
        bond_value = 0
        for bond in self.bonds:
            bond_value += 0
 
        cash_value = self.calculate_cash_value(date)
 
        return bond_value + cash_value
 
 
# geen opgelopen rente in cashmutatie
```
