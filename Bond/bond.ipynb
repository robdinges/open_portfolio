```python
import numpy as np
import numpy_financial as npf
from datetime import datetime

class Bond:
    def __init__(self, face_value, coupon_rate, current_price, issue_date, maturity_date, payment_frequency):
        self.face_value = face_value
        self.coupon_rate = coupon_rate
        self.current_price = current_price
        self.issue_date = datetime.strptime(issue_date, "%Y-%m-%d")
        self.maturity_date = datetime.strptime(maturity_date, "%Y-%m-%d")
        self.payment_frequency = payment_frequency
        self.cash_flows = self._generate_cash_flows()

    def _generate_cash_flows(self):
        """Genereert een lijst van kasstromen van de obligatie."""
        years_to_maturity = (self.maturity_date - self.issue_date).days / 365
        total_payments = int(years_to_maturity * self.payment_frequency)
        cash_flows = [-self.current_price]  # Negatieve kasstroom voor de huidige prijs (investering)
        
        # Voeg couponbetalingen toe
        for _ in range(total_payments - 1):  # Alle behalve de laatste
            cash_flows.append(self.face_value * self.coupon_rate / self.payment_frequency)
        
        # Voeg laatste couponbetaling plus nominale waarde toe
        cash_flows.append(self.face_value * self.coupon_rate / self.payment_frequency + self.face_value)
        
        return cash_flows

    def calculate_ytm(self):
        """Bereken de yield-to-maturity (YTM) van de obligatie."""
        try:
            ytm = npf.irr(self.cash_flows)
            return ytm * self.payment_frequency  # Jaarlijkse YTM
        except:
            return None

# Voorbeeldgebruik
bond = Bond(face_value=1000, coupon_rate=0.033, current_price=990, issue_date="2020-01-01", maturity_date="2030-01-01", payment_frequency=1)

ytm = bond.calculate_ytm()
if ytm is not None:
    print(f"Yield-to-Maturity: {ytm:.2%}")
else:
    print("Yield-to-Maturity: Berekening mislukt of niet beschikbaar.")
```
